```python
from icrawler.builtin import GoogleImageCrawler
import os

# Directory to save random (non-MRI) images
save_dir = 'random_images'
os.makedirs(save_dir, exist_ok=True)

# Keywords that do NOT resemble MRI at all
keywords = [
    'mountains', 'street photography', 'cats', 'dogs', 'cars', 'food', 
    'abstract art', 'people', 'landscape', 'buildings'
]

# Images per keyword (adjust for 2000 total)
images_per_keyword = 200

crawler = GoogleImageCrawler(storage={'root_dir': save_dir})

for keyword in keywords:
    print(f'Downloading: {keyword}')
    crawler.crawl(keyword=keyword, max_num=images_per_keyword)
```

```python
from icrawler.builtin import DuckDuckGoImageCrawler
import os

save_dir = 'random_images'
os.makedirs(save_dir, exist_ok=True)

keywords = [
    'nature', 'cars', 'animals', 'people', 'food', 'buildings',
    'technology', 'sports', 'abstract', 'street art'
]

images_per_keyword = 250  # 10*250 = 2500 total

crawler = DuckDuckGoImageCrawler(storage={'root_dir': save_dir})

for keyword in keywords:
    print(f'Downloading images for: {keyword}')
    crawler.crawl(keyword=keyword, max_num=images_per_keyword)
```

```python
from icrawler.builtin import BingImageCrawler
import os

save_dir = 'random_images'
os.makedirs(save_dir, exist_ok=True)

keywords = [
    'nature', 'cats', 'dogs', 'buildings', 'street art', 
    'food', 'technology', 'toys', 'cars', 'abstract art'
]

images_per_keyword = 200  # 10 x 200 = 2000 total

crawler = BingImageCrawler(storage={'root_dir': save_dir})

for keyword in keywords:
    print(f"Downloading images for: {keyword}")
    crawler.crawl(keyword=keyword, max_num=images_per_keyword)
```

```python
import requests
import os
from tqdm import tqdm

save_dir = "random_images"
os.makedirs(save_dir, exist_ok=True)

num_images = 2000

for i in tqdm(range(num_images), desc="Downloading images"):
    try:
        url = f"https://picsum.photos/512/512"  # 512x512 random image
        response = requests.get(url, timeout=5)

        # Save image
        with open(os.path.join(save_dir, f"random_{i+1}.jpg"), "wb") as f:
            f.write(response.content)
    except Exception as e:
        print(f"Failed to download image {i+1}: {e}")
```
