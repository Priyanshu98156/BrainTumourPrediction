```python
from icrawler.builtin import GoogleImageCrawler
import os

# Directory to save random (non-MRI) images
save_dir = 'random_images'
os.makedirs(save_dir, exist_ok=True)

# Keywords that do NOT resemble MRI at all
keywords = [
    'mountains', 'street photography', 'cats', 'dogs', 'cars', 'food', 
    'abstract art', 'people', 'landscape', 'buildings'
]

# Images per keyword (adjust for 2000 total)
images_per_keyword = 200

crawler = GoogleImageCrawler(storage={'root_dir': save_dir})

for keyword in keywords:
    print(f'Downloading: {keyword}')
    crawler.crawl(keyword=keyword, max_num=images_per_keyword)
```

```python
from icrawler.builtin import DuckDuckGoImageCrawler
import os

save_dir = 'random_images'
os.makedirs(save_dir, exist_ok=True)

keywords = [
    'nature', 'cars', 'animals', 'people', 'food', 'buildings',
    'technology', 'sports', 'abstract', 'street art'
]

images_per_keyword = 250  # 10*250 = 2500 total

crawler = DuckDuckGoImageCrawler(storage={'root_dir': save_dir})

for keyword in keywords:
    print(f'Downloading images for: {keyword}')
    crawler.crawl(keyword=keyword, max_num=images_per_keyword)
```
